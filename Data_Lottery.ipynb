```python
import pandas as pd

def inspect_number():
  df = df_combined
  number = input("Enter a number: ")
  """
  This function inspects a number against a DataFrame to find matches, displays
  results as a DataFrame, including the matching numbers, and checks if the
  first 3 and last 3 digits match specific columns.

  Args:
    df: The DataFrame to inspect.
    number: The number to inspect.

  Returns:
    A DataFrame containing the date, frequency, and matching numbers for
    full, first 3 digits, and last 3 digits matches.
  """

  # Convert the number to a string
  number = str(number)

  # Check if the number matches any values in the '6_number' column
  matches = df[df['6_number'] == number]

  # Check if the first 3 numbers match any values in the 'First 3 numbers' column
  first_3_matches = df[df['First 3 numbers'] == number[:3]]

  # Check if the last 3 numbers match any values in the 'last 3 numbers' column
  last_3_matches = df[df['last 2 numbers'] == number[-3:]]

  if not matches.empty or not first_3_matches.empty or not last_3_matches.empty:

    # Get the dates, frequency, and matching numbers for the 6-digit matches
    dates = matches['date'].values.tolist()
    freq = len(matches)
    matched_numbers = matches['6_number'].values.tolist()

    # Get the dates, frequency, and matching numbers for the first 3-digit matches
    first_3_dates = first_3_matches['date'].values.tolist()
    first_3_freq = len(first_3_matches)
    first_3_matched_numbers = first_3_matches['6_number'].values.tolist()

    # Get the dates, frequency, and matching numbers for the last 3-digit matches
    last_3_dates = last_3_matches['date'].values.tolist()
    last_3_freq = len(last_3_matches)
    last_3_matched_numbers = last_3_matches['6_number'].values.tolist()

    # Create a dictionary for the DataFrame data
    data = {
        "Date(s) - 6 digits": [', '.join(dates)],
        "Frequency - 6 digits": [freq],
        "Matched Numbers - 6 digits": [', '.join(matched_numbers)],
        "Date(s) - First 3 digits": [', '.join(first_3_dates)],
        "Frequency - First 3 digits": [first_3_freq],
        "Matched Numbers - First 3 digits": [', '.join(first_3_matched_numbers)],
        "Date(s) - Last 3 digits": [', '.join(last_3_dates)],
        "Frequency - Last 3 digits": [last_3_freq],
        "Matched Numbers - Last 3 digits": [', '.join(last_3_matched_numbers)],
    }

    # Return the results as a DataFrame
    return pd.DataFrame(data)

  else:
    return "Your number did not match any entries in the DataFrame."
```
